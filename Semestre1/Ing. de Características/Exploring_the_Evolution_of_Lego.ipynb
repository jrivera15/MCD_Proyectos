```python
# Se importan librerías
import pandas as pd
import numpy as np
```

```python
# Se cargan los datos
df_lego_sets_raw = pd.read_csv("datasets/lego_sets.csv")
df_parent_themes_raw = pd.read_csv("datasets/parent_themes.csv")
```

```python
# Este dataset presenta valores perdidos en algunas variables
df_lego_sets_raw.info()
```

```python
df_lego_sets_raw.head()
```

```python
# Dado que la variable *set_num* es la que identifica a cada observación como única, 
# no puede ser nula ni estar duplicada, por lo tanto se eliminan las observaciones 
# con valor nulo para esta variable.
# Los resultados son guardados en nuevo dataframe

df_lego_sets = df_lego_sets_raw.dropna(subset=['set_num'])
df_lego_sets.rename(columns = {'name':'set_name'}, inplace = True)
df_lego_sets.info()
```

```python
# Este dataset no contiene valores perdidos
df_parent_themes_raw.info()
```

```python
df_parent_themes_raw.head()
```

```python
# Se toman todas las observaciones que no son duplicados. 
# El dataset no contiene duplicados.

df_lego_sets[~df_lego_sets.duplicated()]
```

```python
# Se toman todas las observaciones que no son duplicados. 
# El dataset no contiene duplicados.

df_parent_themes = df_parent_themes_raw[~df_parent_themes_raw.duplicated()]
df_parent_themes
```

```python
# Pregunta 1. ¿Qué porcentaje de todos los sets con licencia lanzados alguna vez 
# tenían el tema de Star Wars? Guarde su respuesta como una variable the_force 
# en forma de número entero (por ejemplo, 25).
# -------------------------------------------------------------------------------

# Se hace un merge de ambos dataframes para saber cuales son los sets con licencia
df_merged = pd.merge(
    df_lego_sets, 
    df_parent_themes,
    how="left",
    left_on="parent_theme",
    right_on="name"
)

df_all_licensed = df_merged[df_merged['is_licensed']]
df_all_licensed

# Se toman todos los sets con licencia que sean de Star Wars
df_star_wars_themed = df_all_licensed[df_all_licensed['parent_theme'] == 'Star Wars']
df_star_wars_themed

# Se calcula la proporción
the_force = int((len(df_star_wars_themed) / len(df_all_licensed))*100)
the_force
```

```python
#Pregunta 2. ¿En qué año Star Wars no fue el tema con licencia más popular 
# (en términos de número de sets lanzados ese año)? Guarde su respuesta como 
# una variable new_era en forma de número entero (por ejemplo, 2012).
#----------------------------------------------------------------------

# Se agrupan los sets por año, se toma el nombre del set y se calcula la moda 
# es decir, el nombre del set con más fecuencia
df_most_popular_themes_by_year = df_all_licensed.groupby('year')[['parent_theme']].agg(pd.Series.mode).reset_index()
df_most_popular_themes_by_year

# Se toma el set diferente a Star Wars y se toma el año
df_not_starwars = df_most_popular_themes_by_year[df_most_popular_themes_by_year['parent_theme'] != 'Star Wars']
new_era = int(df_not_starwars['year'])
new_era
```
